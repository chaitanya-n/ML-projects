```python
import os
#from dotenv import find_dotenv, load_dotenv
#from langchain import OpenAI
from transformers import pipeline
from langchain import PromptTemplate, LLMChain
```

```python
# function that uses Huggingface tasks to generate text output from an image input
def image2text(url):
    image_to_text = pipeline("image-to-text", model = "Salesforce/blip-image-captioning-base")
    text = image_to_text(url)[0]["generated_text"]
    #print(text)
    return text
```

```python
#llm - use an opensource LLM to cook up a story for the text that was generated by the "image2text" module.
def generate_story(scenario):
    messages = [{"role": "user", "content": scenario},]
    pipe = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B")
    print("Hello")
    result = pipe(messages)
    #print(story)
    return result
```

```python
# Test output
scenario = image2text("my pic.jpg")
print(scenario)
generate_story(scenario)
```
