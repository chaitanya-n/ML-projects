```python
import os
import streamlit as st
#from dotenv import find_dotenv, load_dotenv
#from langchain import OpenAI
from transformers import pipeline
from transformers import set_seed
#from langchain import PromptTemplate, LLMChain
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# function that uses Huggingface tasks to generate text output from an image input
def image2text(url):
    image_to_text = pipeline("image-to-text", model = "Salesforce/blip-image-captioning-base")
    text = image_to_text(url)[0]["generated_text"]
    #print(text)
    return text
```

```python
#llm - use gpt2 LLM to cook up a story for the text that was generated by the "image2text" module.

def generate_story(scenario):
   pipe = pipeline("text-generation", model="gpt2")
   set_seed(42)
   result = pipe(scenario, max_length = 50, num_return_sequences=5)[1]["generated_text"]
   return result
```
